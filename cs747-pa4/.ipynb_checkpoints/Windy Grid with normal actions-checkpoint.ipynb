```python
import numpy as np
```

```python
initial_pos=[0,3]
final_pos=[7,3]
actions=[0,1,2,3]
max_x=9
max_y=6
#0=up,1=down,2=left,3=right
wind=[0,0,0,1,1,1,2,2,1,0]
np.random.seed(0)
```

```python
def environment(current_pos,a):
    x=current_pos[0]
    y=current_pos[1]
    if a==0:
        y+=1
    elif a==1:
        y-=1
    elif a==2:
        x-=1
    elif a==3:
        x+=1
    if x<0:
        x=0
    if x>max_x:
        x=max_x
    y+=wind[x]
    if y<0:
        y=0
    if y>max_y:
        y=max_y
    return [x,y]
def epsilon_greedy(q,s):
    toss= np.random.choice([0,1],p=[0.1,0.9])
    if toss == 0:
        return np.random.choice(actions)
    elif toss ==1:
        return np.argmax(q[s[0]][s[1]])      
```

```python
q=np.random.rand(max_x+1,max_y+1,len(actions))
q[final_pos[0]][final_pos[1]]=0.0
```

```python
q
```

```python
s=initial_pos
a=epsilon_greedy(q,s)
i=0
while(True):
    i+=1
    print(i)
    s_p=environment(s,a)
    a_p=epsilon_greedy(q,s_p)
    #print(s,a,s_p,a_p)
    g=0.5*(-1+q[s_p[0]][s_p[1]][a_p]-q[s[0]][s[1]][a])
    
    q[s[0]][s[1]][a]+=g
    s=s_p
    a=a_p
    if s == final_pos:
        break
    
    
    
```

```python
time=np.zeros(170)
for episode in range(170):
    s=initial_pos
    a=epsilon_greedy(q,s)
    i=0
    while(True):
        i+=1
        #print(i)
        s_p=environment(s,a)
        a_p=epsilon_greedy(q,s_p)
        #print(s,a,s_p,a_p)
        g=0.5*(-1+q[s_p[0]][s_p[1]][a_p]-q[s[0]][s[1]][a])
        q[s[0]][s[1]][a]+=g
        s=s_p
        a=a_p
        if s == final_pos:
            break
    time[episode]=i
    
    
```

```python
time_steps=np.cumsum(time)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(time_steps,range(170))
```
